We create some functions in order to get the location of the IP from 2 different sources.
The main one will be from DataBase but if doesn't exist, we will use a web source tha is called *freegeoip.net*

The DDBB has IPs in decimal format so we also need to create functions that convert IPs into decimal and against.

````python
def IP_to_decimal(ip):
    return int(netaddr.IPAddress(ip))

def decimal_to_IP(decimal):    
    return str(netaddr.IPAddress(decimal))

def getLocation_fromweb(ipAddress):
    response = requests.get("http://freegeoip.net/json/"+ipAddress)    
    responseJson = json.loads(response.text)
    return responseJson#.get("country_name")

def getLocation_fromDDBB(ip_decimal, ddbb):
    responseJson = dict()
    default_lat = 0
    default_long = 0
    default_country = '-'
    try:
        responseJson['latitude'] = ddbb['latitude'][(ddbb['range_start'] < ip_decimal) & 
                                                (ddbb['range_end'] > ip_decimal)].item()
    except ValueError:
        responseJson['latitude'] = default_lat
    try:
        responseJson['longitude'] = ddbb['longitude'][(ddbb['range_start'] < ip_decimal) & 
                                                (ddbb['range_end'] > ip_decimal)].item()
    except ValueError:
        responseJson['longitude'] = default_lat
    
    try:
        responseJson['country'] = ddbb['country'][(ddbb['range_start'] < ip_decimal) & 
                                                (ddbb['range_end'] > ip_decimal)].item()
    except:
        responseJson['country'] = default_country
    return responseJson

def ip_to_location(ip_list, geo_columns, kind = 'web', ddbb = ''):    
    location_dict = dict()
    read_ips = list()
    
    for ip in ip_list:
        if ip not in read_ips and not pd.isnull(ip):            
            if kind == 'web':
                location = getLocation_fromweb(ip)
            elif kind == 'ddbb':
                location = getLocation_fromDDBB(IP_to_decimal(ip), ddbb)
            location_dict[ip] ={geo_columns['lat']: location['latitude'],
                                geo_columns['long']: location['longitude'],
                                geo_columns['count']: location['country']}
            read_ips.append(ip)
    return location_dict
````

#### From CSV to LIBSVM format (only for numerical data)

For all models in spark is needed that training files are in *libsvm* format. Due to that, we will need to use the next functions

````python
import sys
import csv
from collections import defaultdict

# Source: https://github.com/zygmuntz/phraug/blob/master/csv2libsvm.py

def construct_line( label, line ):
    new_line = []
    if float( label ) == 0.0:
        label = "0"
    new_line.append( label )

    for i, item in enumerate( line ):
        if item == '' or float( item ) == 0.0:
            continue
        new_item = "%s:%s" % ( i + 1, item )
        new_line.append( new_item )
    new_line = " ".join( new_line )
    new_line += "\n"
    return new_line

def csv2libsvm(inputfile, outoutfile, labels=0,headers=0):
    """
    Convert CSV file to libsvm format. Works only with numeric variables.
    Put -1 as label index (label) if there are no labels in your file.
    Expecting no headers. If present, headers can be skipped with headers == 1.
    
    Convert CSV to the LIBSVM format. If there are no labels in the input file,
    specify label index = -1. If there are headers in the input file, specify skip headers = 1.
    """

    input_file = inputfile
    output_file = outoutfile

    try:
        label_index = int( labels )
    except IndexError:
        label_index = 0

    try:
        skip_headers = headers
    except IndexError:
        skip_headers = 0

    i = open( input_file, 'r' )
    o = open( output_file, 'w' )

    reader = csv.reader( i , delimiter = '#')

    if skip_headers:
        headers = next(reader)
    
    for line in reader:
        if label_index == -1:
            label = '1'
        else:
            label = line.pop( label_index )

        new_line = construct_line( label, line )
    
        o.write( new_line )
````

````python
def randomforest(data):
    # Index labels, adding metadata to the label column.
    # Fit on whole dataset to include all labels in index.
    labelIndexer = StringIndexer(inputCol="label", outputCol="indexedLabel").fit(data)

    # Index labels, adding metadata to the label column.
    # Fit on whole dataset to include all labels in index.
    labelIndexer = StringIndexer(inputCol="label", outputCol="indexedLabel").fit(data)

    # Automatically identify categorical features, and index them.
    # Set maxCategories so features with > 4 distinct values are treated as continuous.
    featureIndexer =\
        VectorIndexer(inputCol="features", outputCol="indexedFeatures", maxCategories=4).fit(data)

    # Split the data into training and test sets (30% held out for testing)
    (trainingData, testData) = data.randomSplit([0.7, 0.3])

    # Train a RandomForest model.
    rf = RandomForestClassifier(labelCol="indexedLabel", featuresCol="indexedFeatures", numTrees=100)

    # Convert indexed labels back to original labels.
    labelConverter = IndexToString(inputCol="prediction", outputCol="predictedLabel",
                                   labels=labelIndexer.labels)

    # Chain indexers and forest in a Pipeline
    pipeline = Pipeline(stages=[labelIndexer, featureIndexer, rf, labelConverter])

    # Train model.  This also runs the indexers.
    model = pipeline.fit(trainingData)

    # Make predictions.
    predictions = model.transform(testData)

    # Select example rows to display.
    predictions.select("predictedLabel", "label", "features").show(5)

    # Select (prediction, true label) and compute test error
    evaluator = MulticlassClassificationEvaluator(
        labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")
    accuracy = evaluator.evaluate(predictions)
    print("Test Error = %g" % (1.0 - accuracy))

    rfModel = model.stages[2]
    print(rfModel)  # summary only
````

We plot all flows together from one TLS and one service. First we need some data frame reshapes

````python
def extract_flows(df, service_filter, TLS_filter, column_togroup, column_toplot):
    flow_lenght_column_name = 'flow_lenghts'

    #Filter conditions from interesting info
    c1 = df['COMPA_analyzed_service'] == service_filter
    c2 = df['TLS_SNI_CLH'] == TLS_filter

    # Filter info of a service anda a TLS
    flows_filtered = df[conjunction(c1,c2)]
    # Get the flow_ids
    flowsid_to_plot = flows_filtered[column_togroup].unique().tolist()


    init_list= ['None']
    flows_df= pd.DataFrame({'init': init_list})

    for flow_id in flowsid_to_plot:
        column_name = 'flow_' + str(flow_id)

        # Get lenghts of packets of one flow    
        list_of_lengths = flows_filtered.loc[flows_filtered[column_togroup] == flow_id, column_toplot].tolist()

        # Build the dataframe
        new_flow_df = pd.DataFrame({column_name: list_of_lengths})
        flows_df = pd.concat([flows_df,new_flow_df], axis = 1)

    # Revove init column
    flows_df = flows_df.iloc[:,1:]
    # Set Nan to 0 values
    nan_values_fill = 0
    flows_df = flows_df.fillna(nan_values_fill)
    return flows_df

def df_to_line(flows_df):
    flows_df_line= pd.DataFrame({'flow_id': 'None', 'length': ['None']})
    colum_list = flows_df.columns.tolist()
    for c in colum_list:
        # Get the column of each flow    
        lenght_col = flows_df.iloc[:,colum_list.index(c)]

        # Build the new dataframe column
        new_flow_df = pd.DataFrame({'flow_id': c, 'length':lenght_col})

        # Concat with previous data
        frames = [flows_df_line,new_flow_df]
        flows_df_line = pd.concat(frames)

    # Remove init row
    flows_df_line = flows_df_line.iloc[1:,:]

    flows_df_line = flows_df_line.reset_index(drop=False)
    flows_df_line = flows_df_line.rename(columns = {'index': 'TimePoint'})
    return flows_df_line

def dataframe_convert_toplot(COMPA_TCP_TLS_filtered, service_filter, TLS_filter, column_togroup, column_toplot):
    flows_df = extract_flows(COMPA_TCP_TLS_filtered, service_filter, TLS_filter, column_togroup, column_toplot)
    flows_df_line = df_to_line(flows_df)
    return flows_df_line
````

Finally we plot the flows together in order to find patterns

````python
def plot_flow_patterns(flows_df_line, service, server):
    fig, ax = plt.pyplot.subplots(figsize=(8,6))
    flows_df_line.groupby('flow_id').plot(x = 'TimePoint', y = 'length',kind='line',  ax=ax, legend = False)
    total_flows = flows_df_line.groupby('flow_id').count().iloc[1,1]
    ax.set_xlabel('Number of packet in the flow')
    ax.set_ylabel('Packet length')
    ax.set_title('Flow patterns of SERVICE: ' + service + '\nSERVER: ' +\
                 server + '\nNUMBER OF FLOWS:' +  str(total_flows))
````

Now the target is to create a dataframe with all flows and same flow length. The first column must be the TLS label. The rest of columns will be filled with each packet length of the flow.

In order to get this idea, first we have to extract all the flows and concat all of them, filling the flow tails with 0. First each column will be a flow, with the column name the TLS label

````python
# Function to change all the column names of a dataframe for a given name (TLS label)
def change_column_names(df, new_colname):
    numcols = df.shape[1]
    new_colnames = [new_colname]*numcols
    old_colnames = list(df)
    new_names_dict = {i:j for i,j in zip(old_colnames,new_colnames)}
    new_df = df.rename(columns=new_names_dict, inplace=False)
    return new_df
````

We will set a limit for the flow min and max length. Ir order to reach that point, we will extract all flows, we will plot the flow density and we will decide the limits

````python
# Extract the flows
column_togroup = 'TCP_FlowID'
column_toplot = 'TCP_TCPLen'

frames = []
min_rows = 10
max_rows = 50
factor_remover = 4
lim_to_cut = max_rows * factor_remover

# First, get all the flows
for service in TCP_FlowIDs:
    print('\nService: ', service)
    # Extract the TLSs of one service
    TLS_keys = [k for k in TCP_FlowIDs[service].keys()]
    for TLS in TLS_keys:        
        # Extract all flows of the TLS
        flows = extract_flows(COMPA_TCP_TLS_filtered, service, TLS, column_togroup, column_toplot)
        max_flow_length = flows.shape[0]
        # We avoid flows that exceeds the limit
        if max_flow_length <= min_rows or max_flow_length > lim_to_cut:
            print('Removed service:', service, '|| with TLS:', TLS, '|| and flow in packtes:', max_flow_length)            
        elif max_flow_length <= max_rows:
            # Change the column names
            flows = change_column_names(flows, TLS)
            # Add the flows
            frames.append(flows)
        elif max_flow_length <= lim_to_cut:
            flows = flows.iloc[0:max_rows,:]
            flows = change_column_names(flows, TLS)
            frames.append(flows)
            print('Cut service:', service, '|| with TLS:', TLS, '|| and flow in packtes:', max_flow_length)
        
        
# We concat all dataframes
print('\nMixing frames..')        
all_flows = pd.concat(frames, axis = 1)
print('Finished')
````

````python
def collect_flow_lengths(df):
    col_names = list(df)
    col_lengths = []
    ncol=0
    for col in col_names:
        col_lengths.append(df.iloc[:,ncol].count())        
        ncol+=1
    return col_lengths
````

#### DTW + Agglomerative clustering

The goal is to build clusters based on serie similarity. These clusters will be constructed with flows from different services. After creating the model, we will validate it with the same data, in order to know the percentage of each service inside each cluster.

The first step is to extract a matrix of distances among flows.

````python
# Remove label from dataset
all_flows_t_nolab = all_flows_t.iloc[:,1:]
````

Now we calculate the cost matrix among all flows

````python
######## REDUCE DATASET, ONLY FOR CODING
initlines = 500
num_lines = 50
end_lines = initlines + num_lines

all_flows_t_nolab = all_flows_t_nolab.iloc[initlines:end_lines,:]
all_flows_t_nolab = all_flows_t_nolab.reset_index(drop=True)
all_flows_t_nolab
````

````python
# Extract num of flows
tot_num_flows = all_flows_t_nolab.shape[0]
# Initialize cost_matrix
matrix_cost=np.zeros((tot_num_flows, tot_num_flows))
# Calculate cost matrix
start_time = time.time()
print( 'Beggining DTW' )
start_time = time.time()
for flow in range(0, tot_num_flows):    
    print('Calculating dtw for flow number: ', flow)
    print('Lapsed time: %s seconds' % (time.time() - start_time))
    for f_rest in range(flow, tot_num_flows):        
        print('Cost with flow: ', f_rest)
        if flow!=f_rest:
            x=all_flows_t_nolab.iloc[flow,:]
            y=all_flows_t_nolab.iloc[f_rest,:]
            dist, cost, path = dtw(x, y)

            matrix_cost[flow][f_rest]=np.sum(cost[path[0]][path[1]])
            matrix_cost[f_rest][flow]=np.sum(cost[path[0]][path[1]])

print("---DTW %s seconds ---" % (time.time() - start_time))
````

We calculate line_cost and mean_cost for every line of costs

````python
line_cost = []
mean_cost = []
for k in range(tot_num_flows):
    line_cost.append(matrix_cost[k])
    mean_cost.append(np.mean(matrix_cost[k])) 
````

We create the clusters

````python
st_time = time.time()
# ORIGINAL
# The original code give some options to calculate different amount of clusters
# n_clusters = []
# n_clusters = get_n_clusters_array(len(line_cost))
#
# We are going to get number of clusters from number of flows
n_clusters = tot_num_flows
n_clusters = 5
````

````python
# ORIGINAL
# Here it is fitted the model for every number of cluster
# labels_clust=[]
# for n in n_clusters:
#     average_linkage = AgglomerativeClustering(n_clusters=n, linkage='average')
#     average_linkage.fit(line_cost)
#     labels_clust.append(map(int,average_linkage.labels_.tolist()))
# print labels_clust
#
# We calculate the fit for the given number of clusters
labels_clust=[]
average_linkage = AgglomerativeClustering(n_clusters=n_clusters, linkage='average', affinity = 'euclidean')
average_linkage.fit(line_cost)
# The mapping below is for asigning labels to clusters
labels_clust.append(map(int,average_linkage.labels_.tolist()))
````

````python
# THE ORIGINAL
# num_labels=[]
# for num in range(0,len(labels_clust)):

#     nums_l=[]
#     for cl in range(0,n_clusters[num]):

#         nums_l.append(labels_clust[num].count(cl))
#     num_labels.append(nums_l)
# print(num_labels)
# print("---Clustering %s seconds ---" % (time.time() - st_time))

# Here it is counting the produced labels
num_labels=[]
for num in range(0,len(labels_clust)):
    nums_l=[]    
    for cl in range(n_clusters):
        nums_l.append(labels_clust.count(cl))
    num_labels.append(nums_l)
print(num_labels)
print("---Clustering %s seconds ---" % (time.time() - st_time))


# PROBAR ELIMINANDO FLUJOS CON MUCHOS CEROS
````
